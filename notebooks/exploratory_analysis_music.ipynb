# Exploratory Data Analysis: Музыкальные треки

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('../data/spotify_tracks.csv')
df['duration_min'] = df['duration_ms'] / 60000
df.head()
```

## Топ-10 жанров по количеству треков

```python
genre_counts = df['genre'].value_counts().head(10)
sns.barplot(x=genre_counts.values, y=genre_counts.index)
plt.title('Топ-10 жанров по количеству треков')
plt.xlabel('Количество треков')
plt.show()
```

## Распределение популярности треков

```python
sns.histplot(df['popularity'], bins=20, kde=True)
plt.title('Распределение популярности треков')
plt.xlabel('Популярность')
plt.show()
```

## Energy vs Popularity

```python
sns.scatterplot(data=df, x='energy', y='popularity', alpha=0.3)
plt.title('Energy vs Popularity')
plt.show()
```

## Valence vs Danceability

```python

sns.scatterplot(data=df, x='valence', y='danceability', alpha=0.3)
plt.title('Valence vs Danceability')
plt.show()
```

## Длительность треков

```python
sns.histplot(df['duration_min'], bins=20, kde=True)
plt.title('Распределение длительности треков (в минутах)')
plt.xlabel('Длительность (мин)')
plt.show()
```

## Мажор vs Минор

```python
mode_labels = {0: 'Minor', 1: 'Major'}
mode_popularity = df.groupby('mode')['popularity'].mean().rename(index=mode_labels)
sns.barplot(x=mode_popularity.index, y=mode_popularity.values)
plt.title('Средняя популярность: мажор vs минор')
plt.ylabel('Популярность')
plt.show()
```

## Выводы
- Популярность треков распределена неравномерно
- Энергичные треки чаще бывают популярнее
- Жанры `pop` и `rock` — самые многочисленные
- В среднем треки мажора немного популярнее, чем минорные
